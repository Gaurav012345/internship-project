```python
from statsmodels.formula.api import ols
```

```python
#df.rename(columns={'2B':'2_B','3B':'3_B'},inplace = True)
df.rename(columns={'2B':'B_2','3B':'B3'},inplace = True)
```

```python
df.columns
```

```python
m = ols('W ~ SV+CG+BB+R+ERA+B_2',data=df).fit()
```

```python
m.summary()
```

```python
import statsmodels.api as sm
```

```python
sm.stats.anova_lm(m,type=2)
```

```python
pred = m.predict(df[['SV','CG','BB','R','ERA','B_2']])
```

```python
res = m.resid
```

```python
plt.plot(df['W'],pred)
plt.show()
```
